```python
import pandas as pd
```

```python
alabama = "Resources/alabama_District_all.csv"
```

```python
alabama_df = pd.read_csv(alabama, encoding = "ANSI")
```

```python
alabama_df.head()
```

```python
# join together columns: "Topic", "Subject", and "Title" inorder to avoid 
# previous problems with duplicate column names after transform
alabama_df['Title'] = alabama_df[alabama_df.columns[0:3]].apply(
    lambda x: '-'.join(x.dropna().astype(str).astype(str)),
    axis=1)
#answer from StactOverflow:
#https://stackoverflow.com/questions/33098383/merge-multiple-column-values-into-one-column-in-python-pandas

alabama_df.head()
```

```python
#pull column names into list
cols = alabama_df.columns.tolist()

#establish matching criteria for column names we want
matchers = ["Title", "Estimate"]

#make list of columns including only column names that have matchers in them
#we'll use that to make a df with only the hard district data we want, without the MOE
new_cols = [s for s in cols if any(xs in s for xs in matchers)]
```

```python
#test
print(cols)
print(new_cols)
```

```python
# make new df with only 'new_cols'
alabama_df_districts = alabama_df[new_cols].copy()
#test
alabama_df_districts.head()
```

```python
#prep for transform so rename "Title" to "state_dist" and set it as the index 
# this will also allow for SQL joins
alabama_df_districts = alabama_df_districts.rename(columns={"Title":"state_dist"})
alabama_df_districts = alabama_df_districts.set_index('state_dist')
#344 rows
alabama_df_districts
```

```python
# Drop NAN and get rid of comments at the bottom
alabama_df_districts.dropna(inplace=True)
# 303 rows

#there is clearly a loss of data in the business sectors area, specifically Utilities 
#not sure exactly how to handle that.

alabama_df_districts
```
